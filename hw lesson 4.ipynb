```python
phone_number = input('Введите Ваш телефонный номер:')

def normalize_phone(a_function_to_decorate):
    def the_wrapper_around_the_original_function(phone_number):
        if len(phone_number) == 13:
            return phone_number
        elif len(phone_number) == 12:
            print(f'+{phone_number}')
        elif len(phone_number) == 9:
            print(f'+375{phone_number}')
        else:
            print('Вы ввели неверный формат телефона')
    return the_wrapper_around_the_original_function

@normalize_phone 
def print_phone(phone_number): 
    print(phone_number)

print_phone(phone_number)
```
